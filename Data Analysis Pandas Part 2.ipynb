```python
# In this part we will look US minimum wage data available for some years on state level. The data is available on kaggle.
# The data is parsed from the US govt. website and hence have some issues. The encoding is in latin.
import pandas as pd

df = pd.read_csv("/Users/atyagi/Desktop/Python Tutorials/Data-Analysis-in-Python/Dataset/Minimum Wage Data.csv", encoding = "latin")
df.to_csv("/Users/atyagi/Desktop/Python Tutorials/Data-Analysis-in-Python/Dataset/Minimum Wage Data utf-8.csv", encoding = "utf-8")
df = pd.read_csv("/Users/atyagi/Desktop/Python Tutorials/Data-Analysis-in-Python/Dataset/Minimum Wage Data utf-8.csv")
df.head()
```

```python
# One of our interest in this data is to see the different statistic of lowest wage across State.
# To do that we need a new data frame that has lowest wage for each state in seperate columns.
gd = df.groupby("State") # We are creating a groupby object by State variable.
#Now we can do a lot of things on this groupby object.
gd.get_group("Alabama").set_index("Year").head()
# We are seperating the Albama State and then setting Year as index for the data frame.
```

```python
# Now let's create our data frame for each state with lowest wage.
act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"): # groupby object has 2 elements. one is name of each and the data frame for that particular state.
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["Low.2018"]].rename(columns = {"Low.2018": name})
        # rename is the function used for renaming columns and the arguments are passes as dictionary.
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns = {"Low.2018": name}))

act_min_wage.head()
```

```python
act_min_wage.describe()
```

```python
# Now in the above data we see some of the states have NaN or 0 as minimum wage. 
# We wants to see what columns have all the missing values and what have some missing values. 
df[df["Low.2018"] == 0]["State"].unique()
```

```python
# So we have 15 states that have this issue of 0 minimum wage
df[df["Low.2018"] == 0]["State"].unique().size
```

```python
# Now what we want to do is to convert 0 to NA, otherwise we have to check for 0 and NA seperately.
import numpy as np
act_min_wage = act_min_wage.replace(0, np.NaN)
act_min_wage.head()
```

```python
# Now we can see how many state have how many zero values.
for name in act_min_wage.columns: # .columns just gives a list of all the columns in a data frame
    if act_min_wage[f"{name}"].isna().sum() == 50: # testing what all states have all the missing values
        print(name)
```

```python
# Now in some cases we need to drop columns that have any 0 values and in some cases we just need to drop colums that have all the zero values.
# For eg. correlation matrix
act_min_wage.replace(0,np.NaN).dropna(axis=1).corr().head()
```

```python
# Now let's say we just want to drop columns that have all the zero values
all_zero = []
for name in act_min_wage.columns: # .columns just gives a list of all the columns in a data frame
    if act_min_wage[f"{name}"].isna().sum() == 50: # testing what all states have all the missing values
        all_zero.append(name)

act_min_wage.drop(all_zero, axis = 1).head()
```

```python
# Now we will import the US unemployement data available on kaggle as well.

unemp_county = pd.read_csv("/Users/atyagi/Desktop/Python Tutorials/Data-Analysis-in-Python/Dataset/Unemployment.csv")
unemp_county.head()
```

```python
unemp_county.shape
# The unemployement data is available to us on County, Year and State level and 
# the minimum wage data was on Year and State level. 
# Now we wants to map minimum wage to the unemployement data. One way to do that is to use map function.
```

```python
def get_min_wage(Year, State):
    try:
        return act_min_wage.loc[Year][State] #.loc is used to access a group of rows and columns by label
    except:
        return np.NaN
get_min_wage(1968,"Colorado")
```

```python
# map function takes a lot of time to map the data but works quite efficiently.
unemp_county["min_wage"] = list(map(get_min_wage, unemp_county["Year"], unemp_county["State"]))
```
