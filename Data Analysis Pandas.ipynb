```python
# The warning that we are getting can be solved if we create a copy of the data frame df before using it for further use.
albany_df = df.copy()[df["region"] == "Albany"]
albany_df.set_index("Date", inplace = True)
albany_df.sort_index(inplace = True)
albany_df["price25ma"] = albany_df["AveragePrice"].rolling(25).mean()
```

```python
# The starting values are going to be blank because 25 month average is can't be calculated for them.
albany_df.dropna().head(3) # This will drop all rows that have any NaN value in any of the column.
```

```python
# Now let's try to graph for all the values of region.
# Now to do that first we need a data frame that columns and each column has price MA for each region.
# So we need to filter all the region first and then combine them together over the index.
# before we do that we have an issue that we haven't noticed yet which is that we have duplicates in dates.
# This is because we have more than one value in type column.
df["type"].unique()
```

```python
# So we are going to seperate out the organic type and plot only that.
df = df.copy()[df["type"] == "organic"]
df["Date"] = pd.to_datetime(df["Date"])
df.sort_values(by = "Date", inplace = True, ascending = True)
df.head()
```

```python
# Now we can proceed with creating a data frame that contain MA for each region
graph_df = pd.DataFrame()
for region in df["region"].unique():
    region_df = df.copy()[df["region"] == region]
    region_df.set_index("Date", inplace = True)
    region_df.sort_index(inplace = True)
    region_df[f"{region}_price25ma"] = region_df["AveragePrice"].rolling(25).mean()
    if graph_df.empty:
        graph_df = region_df[[f"{region}_price25ma"]]
    else:
        graph_df = graph_df.join(region_df[f"{region}_price25ma"])
```

```python
graph_df.tail()
```

```python
graph_df.plot(legend = False, figsize = (8,5))
```

```python
# We have some empty space in the beginning because of NA value in MA.
graph_df.dropna().plot(legend = False, figsize = (8,5))
```
